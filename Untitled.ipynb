```python
import pandas as pd
```

```python
ans = {
    'k1': ['A1', 'A1', 'A1', 'A2', 'A3', 'A3', 'A3'],
    'k2': ['B1', 'B1', 'B1', 'B2', 'B2', 'B3', 'B3'],
    'k3': ['C1', 'C2', 'C1', 'C2', 'C3', 'C3', 'C3'], 
    'S' : [1,2,3,4,5,6,7],
    'M' : [2,3,4,5,6,7,8],
}

run = {
    'k1': ['A1', 'A2', 'A3'],
    'k2': ['B1', 'B2', 'B3'],
    'k3': ['C1', 'C2', 'C3'], 
    'S' : [2, 3, 4],
    'M' : [1, 1, 1],
}

fe_list = ['k1', 'k2', 'k3']
v_list = ['S', 'M']

concate_fe = lambda x : '_'.join(x)
```

```python
df_ans = pd.DataFrame(ans)
df_run = pd.DataFrame(run)
```

```python
df_ans
```

```python
df_run
```

```python
df_ans['comb'] = df_ans[fe_list].apply(concate_fe, axis=1)
df_run['comb'] = df_run[fe_list].apply(concate_fe, axis=1)
ans_comb_set = set(df_ans['comb'])
run_comb_set = set(df_run['comb'])
```

```python
ans_comb_dict = {}
run_comb_dict = {}
for comb in sorted(ans_comb_set):
    ans_comb_dict[comb] = df_ans[df_ans['comb']==comb].index
for comb in sorted(run_comb_set):
    run_comb_dict[comb] = df_run[df_run['comb']==comb].index
print(ans_comb_dict)
print(run_comb_dict)

print(ans_comb_dict.keys())
print(run_comb_dict.keys())
```

```python
df_res = pd.DataFrame()
for comb in run_comb_dict.keys():
    if comb in ans_comb_dict.keys():
        # print(comb)
        df_r = df_run.loc[run_comb_dict[comb]].copy()
        df_a = df_ans[df_ans['comb']==comb].copy()
        
        for v in v_list:
            df_a['_'+v] = df_r.iloc[0][v]
            df_a[v+'_MAE'] = df_a[v] - df_a['_'+v]
            df_a[v+'_MAE_mean'] = df_a[v+'_MAE'].mean()
        print(df_a)
        df_res = df_res.append(df_a.copy())
    
    print()
df_res
```

```python
for v in v_list:
    df_res[v+'_MAE'] = df_res[v] - df_res['_'+v]
df_res['S_MAEm'] = df_res['S_MAE'].mean()
df_res
```
