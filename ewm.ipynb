```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('sig.csv')
```

```python
df
```

```python
df['Tx_ewm_t'] =df['100+Tx'].ewm(span=5, adjust=True, min_periods=5).mean()
df['Tx_offset'] = df['Tx_ewm_t'] - df['Tx_ewm_t'].shift(1)
df['Tx_offset_sum'] = df['Tx_offset'].cumsum()
a = list(df['Tx_offset_sum'].dropna())[0]
df['Tx_offset_sum_final'] = df['Tx_offset_sum'] - a
```
